## Arrays

```python
import random

a = list(range(15))
random.shuffle(a)
print(a)
```

```python
b = a[:10]
c = a[10:]

print(b[-2])
```

```python
a.sort()
print(a)
print(b)
print(c)
```

## Matrix

```python
import numpy as np

m = np.random.rand(3, 4) * 100
m = m.astype(int)
print(m)
```

```python
print(m[:2,:])
print(m[:,:1])
```

```python
nm = m[-2:, -2:]
print (nm)
```

## Plotting

```python
import matplotlib.pyplot as plt
x = np.arange(0, 1, 0.1)
y = (2 * x) + 5
print(plt.plot(x,y))
```

```python
y = (x**2) + 5
print(plt.plot(x,y))
```

```python
na = np.random.randn(1000)
plt.hist(na)
```

## Dispersion

```python
N = 50
disp_x = np.random.rand(N)
disp_y = np.random.rand(N)
colors = list(range(50))
area = list(range(50))

print(plt.scatter(disp_x, disp_y, alpha = 0.5, c = colors, s = area))
```
