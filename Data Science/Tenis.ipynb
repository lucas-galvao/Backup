```python
import wget
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#wget.download('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2018.csv')
```

```python
df = pd.read_csv('atp_matches_2018.csv')
#print(df.head(5))
```

## Vencedor valor absoluto

```python
v_total = df.groupby(['surface','winner_name'], as_index = False).count()
v_total = v_total.sort_values(by='tourney_id', ascending=False)
v_total = v_total[['winner_name','surface', 'tourney_id']]
#print(v_total.head(20).plot.bar())
print(v_total.head())
```

## Vencedor por piso

```python
v_surface = df.groupby(['winner_id', 'winner_name', 'surface']).count()
v_surface = v_surface.sort_values(by='tourney_id', ascending=False)
v_surface = v_surface['tourney_id']
#x = v_surface['tourney_id'].values()
```

### Hard

```python
hard = v_surface.filter(like='Hard')
hard
#hard.head(12).plot.bar()
#plt.title('Ganhadores Hard')
#plt.xlabel('Jogadores')
#plt.ylabel('Vitoria')
#plt.show()
```

### Clay

```python
clay = v_surface.filter(like='Clay')
clay = clay.to_frame()
clay.head(12).plot.bar()
plt.title('Ganhadores Hard')
plt.xlabel('Jogadores')
plt.ylabel('Vitoria')
plt.show()
```

### Grass

```python
grass = v_surface.filter(like='Grass')
grass.head(12).plot.bar()
plt.title('Ganhadores Hard')
plt.xlabel('Jogadores')
plt.ylabel('Vitoria')
plt.show()
```
