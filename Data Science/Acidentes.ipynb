```python
tabela = []

with open('acidentes2019.csv', 'r', encoding="utf8") as f:
    tabela = f.readlines()
    
header = tabela[0].split(";")
tabela = tabela[1:]

header
```

```python
bairro = {}

i = header.index("bairro")

for x in tabela:
    x = x.split(";") 
    x = x[i]
    if x in bairro.keys():
        bairro[x] += 1
    else:
        bairro[x] = 1

bairro_ord = dict(sorted(bairro.items(), key=lambda x: x[1], reverse=True))
```

```python
import matplotlib.pyplot as plt

plt.bar(list(bairro_ord.keys())[:17], list(bairro_ord.values())[:17])
plt.xlabel('Bairro')
plt.ylabel('Acidentes')
plt.xticks(rotation=90)
plt.title('Bairro com maior n√∫mero de acidentes em 2019')
plt.show()
```
